The dataframe PPL provides the necessary data for predicting the population trend.

```python
df_Asian = df_PPL.drop(columns=['1990 Total', '2010 Total', '2017 Total'])
df_Asian = df_Asian.drop([0, 1, 2, 4, 5])
df_Asian
```

```python
import matplotlib.pyplot as plt

df_Asian.plot(kind='bar')
plt.xlabel ('Asian Percentage')
```

As you can see, the growth of the Asian population is rapid. Although it seems like it has slow down a little from 2010 to 2017, the growth is still strong comparing to other race.


```python
df_AsianN = df_NPL.drop(columns=['1990 Total', '2010 Total', 'Percentage Change'])
df_AsianN = df_AsianN.drop([0, 1, 2, 4, 5])
df_AsianN = df_AsianN.replace(to_replace ="Asian", 
                 value ="NE Phila")

df_AsianM = df_Asian.drop(columns=['2017 Percentage makeup'])
df_AsianM = df_AsianM.replace(to_replace ="Asian", 
                 value ="Phila")
frames = [df_AsianM, df_AsianN]
Compare = pd.concat(frames)
Compare
```

```python
Compare.plot(kind='bar', x = 'Race')
plt.xlabel ('Asian Percentage')
```

The above comparison shows the Northeast Philadelphia has lower Asian ratio in 1990, but higher Asian ratio in 2010. 
This explains why some Chinese consider the Northeast Philadelphia “Little China Town”. 
The data indicates there is a potential market for the Asian segment.

Result:
Putting the data side by side, we can see the population between New York City and Philadelphia is quite similar.
The percentage of Female population and the student population (age below 18), are very similar, and the Asian segment is growing rapidly, which indicate we do have the “demand” for this type of store.

Based on the search on FourSquare, the whole Philadelphia area lacks the supply of a Chinese Dessert store. The only similar competition will be from those milk/fruit tea store. However, a Chinese Dessert store sells actual dessert, not just drink. The competition from the milk/fruit store can be ignored at this moment.
